## 서브플롯 개념

```python
# 영역을 확보하는 subplot

import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(16,8))

plt.subplot(2,2,1) # 2행 2열로 1개
plt.subplot(2,2,2) # 2행 2열로 2개
plt.subplot(2,2,3) # 2행 2열로 3개
plt.subplot(2,2,4) # 2행 2열로 4개

plt.show()
```

```python
plt.figure(figsize=(16,8))

plt.subplot(211) 
plt.subplot(223) 
plt.subplot(224)

plt.show()
```

```python
plt.figure(figsize=(16,8))

plt.subplot(2,3,(1,2)) 
plt.subplot(2,3,3) 
plt.subplot(2,1,2)

plt.show()
```

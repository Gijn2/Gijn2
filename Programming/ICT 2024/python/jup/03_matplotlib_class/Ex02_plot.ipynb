```python
# 기본적인 그래프

import matplotlib.pyplot as plt
import numpy as np

plt.plot(np.random.randn(100))
plt.plot(np.random.randn(200).cumsum())
```

```python
# 1- 그래프 틀(영역)을 만들기
# 2- 그래프 그리기
# 3- 그래프 보여주기

fig = plt.figure()

ax1 = fig.add_subplot() # 나는 1행 2열 중 1열     [그림]   1행 -1열 -2열
ax2 = fig.add_subplot() # 나는 1행 2열 중 2열

ax1.plot(np.random.randn(100))
ax2.plot(np.random.randn(200).cumsum())

plt.show()
```

```python
# 위와 같은 코딩
fig, ax = plt.subplots(1,2)
ax[0].plot(np.random.randn(100))
ax[1].plot(np.random.randn(200).cumsum())
```

```python
# 간단한 한글처리
from matplotlib import font_manager, rc
import matplotlib.pyplot as plt

plt.rcParams['axes.unicode_minus']=False  # 추가설정 : 폰트를 변경하면 -표시가 ㅁ으로 변경되기에 '-'를 변경하지 않도록 지정
rc('font', family='Malgun Gothic') # 맑음체(malgun) (없다면) 고딕체(gothic)
```

```python
# x축과 y축의 라벨 추가
data=[10,15,19,50,20,30]

#plt.figure() # 사이즈를 변경할 경우, 크기를 잡아준다.
plt.figure(figsize=(16,8))
plt.plot(data);
plt.title('중요한 데이터 차트')
plt.xlabel('시간')
plt.ylabel('값')

plt.show()
```

1- 그래프 틀(영역)을 만들기
   > plt.figure()       
    
2- 그래프 그리기
   > plt.plot(data)     
    
3- 그래프 보여주기
   > plt.show()         

 
 ### *그래프 출력시 객체표시 같은 거 안 출력하게 하려면 뒤에 세미콜론(;) 붙이면 된다*

```python
# y축 데이타에 x축 데이타를 지정

step=['하나요','두식이','석삼','너구리','오징어','육개장']
plt.plot(step,data); # 굳이 따로 안적어도 위의 데이터를 가져올듓
plt.grid()


```

```python
# 속성 : linestyle , color , marker
plt.plot(step,data,linestyle='dashed',color='green',marker='o')
```

```python
# 속성 축약
plt.plot(step,data,'--bo');
```

[참고] 선의 두께와 선의 색을 지정

```python
%matplotlib inline
    
import matplotlib.pyplot as plt
import numpy as np

""" 0. 플롯 스타일 """
plt.style.use('classic') # plot design

x = np.linspace(0,10,100) # 0 ~ 10 사이 수를 100 개로 균등하게 ?
x

plt.plot(np.sin(x)) # sin(x) 함수
```

```python
""" 1. 색상지정 """
plt.plot(np.sin(x-0),color='red');
plt.plot(np.cos(x-1),color='g');
plt.plot(np.tan(x-2),color='#FF9933');
plt.plot(np.sin(x-3),color='chartreuse');
plt.plot(np.sin(x-4),color=(0.5,0.2,0.1));
```

```python
""" 2. 선스타일 """
# plt.plot(np.sin(x), linestyle='solid');
# plt.plot(np.sin(x-1), linestyle='dotted');
# plt.plot(np.sin(x-2), linestyle='dashed');
# plt.plot(np.sin(x*3), linestyle='dashdot');

# 위와 같은 표현
plt.plot(np.sin(x), linestyle='-');
plt.plot(np.sin(x-1), linestyle=':');
plt.plot(np.sin(x-2), linestyle='--');
plt.plot(np.sin(x*3), linestyle='-.');
```

## r(red) g(green) b(blue) c(cyan) m(magenta) y(yellow) k(black) 

```python
""" 3. 선과 스타일을 간결하게 (대략 읽는 수준으로 ) """

plt.plot(np.sin(x+1),'g-')   # solid + green
plt.plot(np.sin(x+2),'--c')# dashed + cyan
plt.plot(np.sin(x+3),'-.k')# dashdot + black
plt.plot(np.sin(x+4),'r:') # dotted + red
```

```python
''' 4. 범례 '''
plt.plot(np.sin(x+1),'g-', label='first')   # solid + green
plt.plot(np.sin(x+2),'--c',label='second')  # dashed + cyan
plt.plot(np.sin(x+3),'-.k',label='third')   # dashdot + black
plt.plot(np.sin(x+4),'r:', label='forth')   # dotted + red

plt.legend(loc=1) # 1 ~ 10 : 대각선 - 상하좌우 - 중앙 (5번 7번 겹침)
```

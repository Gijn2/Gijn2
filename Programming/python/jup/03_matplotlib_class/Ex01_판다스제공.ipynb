### 1. Series 의 plot()

```python
%matplotlib inline
```

```python
from pandas import Series # as pd 라는 별칭을 준다 (대체로)
```

```python
# 판다스 그래프 : Series의 plot() 
s=Series([1.5, 2.3, 0.9], index=['no1','no2','no3'])
s
```

```python
s=Series([1.5, 2.3, 0.9], index=['no1','no2','no3'])

# shift + tab -> default : line
splot = s.plot(kind='bar');

splot.set_xlabel('data2')   # x-line title
splot.set_ylabel('value2')  # y-line title
splot.set_title('sample')   # graph title
```

## 2. DataFrame 그래프

```python
from pandas import DataFrame
import numpy as np

# 데이타프레임에서 시각화 작업
df=DataFrame(np.random.randn(10, 4),    # 난수 10개를 4번
             columns=['level','step','cnt','temp'], 
             index=np.arange(0,100,10))     # np.arange(0,100,10)

df

df.plot() # 꺾은선 그래프
```

```python
df.plot(kind='barh', title='Data Chart', xlabel='x-title', ylabel='y-title') 
# barh : 수평으로 나타나는 bar graph
# shift + tab 으로 읽으면서 위와 같이 내부에 제목을 추가할 수 있다.
```
